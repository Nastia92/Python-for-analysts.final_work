### Скачать следующие данные: kc-house-data и laptop_price

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 6))
plt.hist(df['price'])
plt.title('cost_objects')
plt.xlabel('cost')
plt.ylabel('quantity');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['sqft_living'], bins=20)
plt.title('distribution_residential_quadrature')
plt.xlabel('square_meter')
plt.ylabel('quantity');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['yr_built'], bins=30)
plt.title('Distribution_year_construction')
plt.xlabel('year_construction')
plt.ylabel('quantity');
```

### 2.2 Изучите распределение этажей домов

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Стоимость недвижимости зависит от площади. Чем больше площадь тем выше стоимость

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Самая высокая стоимость у объектов лучшего качества

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution_price_grade');
```

Чем выше класс тем выше стоимость

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution_price_floors');
```

Стоимость зависит от этажности. Выше этаж-больше стоимость

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution_price_view');
```

Стоимость зависит от окружающего вида
